```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
from scipy import integrate, interpolate
```

Let consider a particle in a sequence of potential wells. For instance, an electron in a crystal, where wells are nuclei. We ask ourselves what are the allowed energies of the electron? In particolar, we are interested in the bound states.

```python
x = np.linspace(-5, 5, 1000)
y = np.sum([-np.exp(-(x-a)**2/0.01) for a in range(-4, 5)], axis=0)
plt.figure(figsize=(8,2))
plt.plot(x,y)
```

The ith excited state of the jth well is described as $|i\rangle_j$. In the Linear Combination of atomic orbitals approximation, each state is given by:
$$
|\Psi \rangle = \sum_{ji} |i\rangle_j = \begin{pmatrix}
c_1 \\ 
c_2 \\
c_3\\
c_4 \\
\ldots
\end{pmatrix}
$$
i.e.
$$
\Psi \rangle = c_1 |0\rangle_0 + c_2 |1\rangle_0 + c_3|0\rangle_1 + \ldots
$$
The Hamiltonian can be written n the $i\rangle_j$ basis:
$$
H = T + \sum_i V_i \longrightarrow \langle i'|_{j'} H |i\rangle_j
$$
Let write $H = H_j + V_j$ where $H_j$ is the Hamiltonian just for a $j$ well, i.e. $\hat{V}_j = sum_{m\neq j} V_m$. We set $n = j-j'$ (number of wells apart):
$$
\begin{align}
\langle i'|_{j'} H |i\rangle_j &= \langle i'|_{j'} | E_i + \hat{V}_j |i\rangle_j \\
&= \int \phi^{\star}_{i'}(x-nb)/E_i + \hat{V}(x))\phi_i(x) dx
\end{align}
$$
We need the energy of each well to solve the above integral. $\hat{V}(x)$ can be obtained by the single wells. 

```python
def solve_single_well(x, V, V0):
    N = len(x)
    dx = np.diff(x)[0]
    main_diag = 2*np.ones(N)/dx**2 + V
    off_diag = -np.ones(N-1)/dx**2

    Es, psis = eigh_tridiagonal(main_diag, off_diag, select="v", select_range=(-V0, 0))

    psis = psis/np.sqrt((np.abs(psis)**2).sum(axis=0)*dx)

    return Es, psis.T
```

Finite Square Well:
V0 = 1 eV, 2a (width of the well) = 0.35 nm and well spacing of b = 3*2a.

```python
V0 = 1 # ev
hbar = 6.582112e-16 # eV s 
m = 0.510999e6 / (3e8)**2 # ev s^2 / m^2
a = 0.35e-9 / 2 # m
f = hbar**2 / (2*m*a**2)
# dimensionless
V0 = V0 / f
b = 3*2
```

```python
N = 100000
x = np.linspace(-15,15,N)
```

```python
V = -V0*np.array((x>=-1)*(x<=1), dtype=float)
```

```python
plt.figure(figsize=(6,3))
plt.plot(x, V)
plt.show()
```

```python
Es, psis = solve_single_well(x, V, V0)
```

```python
# to solve the integral, we need to create a function that can compute the wavefunctions 
# at any point x
```

```python
psis.shape
```

```python
# let interpolate the numerical wavefunction
psis_f = [interpolate.interp1d(x, psi, bounds_error=False, fill_value=(0,0)) for psi in psis]
```

Above we are cycling over possible bound states but actually we have just one bound state in this example.

```python
def V(x, V0):
    return -V0*np.array((x>=-1)*(x<=1), dtype=float)
```

```python
# The below formula allows to create the periodic function for V, needed:
plt.plot(x, V((x-2.5)%5-2.5,V0)-V(x, V0))
plt.show()
```

```python
# define hat(v), i.e. the potential from all the wells except the one centered at x = 0
def Vp_f(x, b, V0):
    return V((x+b/2)%b-b/2,V0)-V(x,V0)
```

```python
x_large = np.linspace(-20, 20, 10000)
plt.figure(figsize=(6,3))

plt.plot(x_large, psis_f[0](x_large)**2)
plt.plot(x_large, psis_f[0](x_large-b)**2)
plt.plot(x_large, Vp_f(x_large, b, V0)/V0, color="red")
plt.show()
```

```python
def compute_integral(psis_f, Es, i1, i2, b, V0, n):
    return integrate.quad(
        lambda x: psis_f[i1](x)*psis_f[i1](x-n*b)*(Es[i1]+Vp_f(x, b, V0)),
        -20, 20, points = [n*b for n in range(-6, 6)], limit=1000)[0]
```

```python
N_states = len(Es)
N_states
```

```python
N_wells = 20
H = np.zeros((N_wells*N_states, N_wells*N_states))
```

```python
# now fill off diagonals
# notice that in this case, we have just a single state (ground state) for 
# each well, so the multiplication gives a list where the i element interpolated 
# wavefunction of the j well. If we have two states, then the first two rows are the 
# wavefunctions of the first well, the 3rd and 4th the states of the second well, etc
N_wells*psis_f
```

```python
psis_tot = N_wells*psis_f
Es_tot = N_wells*list(Es)
for i1 in range(H.shape[0]):
    for i2 in range(i1, H.shape[0]):
        well1_idx = i1//N_states
        well2_idx = i2//N_states
        H[i1,i2] = compute_integral(psis_tot, Es_tot, i1, i2, 
                                    b=b, V0=V0, n=well2_idx-well1_idx)
H += H.T
```

```python
for i1 in range(H.shape[0]):
    for i2 in range(i1, H.shape[0]):
        H[i1, i2] = 1
```

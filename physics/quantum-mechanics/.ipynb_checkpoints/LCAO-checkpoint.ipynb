# Linear Combination of Atomic Orbitals (LCAO)

In this notebook, we discuss a popular approximation method to treat molecular states in quantum chemistry, namely the so-called **linear combination of atomic orbitals (LCAO)**. The basic idea behind such approximation consists in defining any molecular orbital $\Psi$ (i.e. the wavefunction of a molecular system) as the superposition of atomic orbitals $\chi_i$:
$$
\Psi = \sum_i c_i \chi_i
$$
where atomic orbitals are wavefunctions centered at the atomic positions and acts as a basis set of functions to expand the molecular state. 
Here, we illustrate this technique by means of a simple but instructive example, namely that of a particle moving in a sequence of potential wells. This system can be considered as a rough approximation of an electron in a metal, where wells are nuclei. In particular, we will try to solve the following problem: *what are the allowed energies for the electron*? 

## Theory

First of all, let briefly discuss LCAO from a theoretical perspective, with a simple example involving just a two-atom system. The Hamiltonian of an electron moving in such diatomic molecule will be:
$$
H = T + V_1 + V_2
$$
where $T$ is the kinetic energy, $V_1$ is the potential energy due to the first nucleus, while $V_2$ is the potential due to the second nucleus. We know that for an electron in a single potential well, we may have different *bound states* of different energies. Using a chemical language, the electron can live in different atomic orbitals (which are theirselves solution to the time-independent Schrodinger equation) of different energy and shape. Let assume to have just one orbital per atom, i.e. we consider just a single basis function centered at each atom position. Additionally, let assume that atoms are sufficiently far apart so that we can neglect any overlapping effect between orbitals. Let call $|1\rangle$ and $|2\rangle$ the atomic orbital of the electron on the first atom and on the second atom respectively. Therefore, we may write:
$$
\begin{gather}
(T+V_1)|1\rangle = \epsilon_0 |1\rangle \\
(T+V_2) |2\rangle = \epsilon_0 |2\rangle 
\end{gather}
$$
The LCAO method consists in searching for a solution which is the linear combination of the atomic orbitals, i.e.
$$
|\Psi\rangle = c_1 |1\rangle + c_2 \rangle
$$
In this context, $|\Psi\rangle$ is called **molecular orbital** because it is an eigenstate of an electron in a diatomic molecule. Assuming that $\langle 1 | 2 \rangle = 0 $, $|\Psi\rangle$ is correctly normalized, providing that $\sum_i |c_i|^2 = 1 $:
$$
\langle \Psi | \Psi \rangle = |c_1|^2 + |c_2|^2 + c_1^{\star}c_2\langle 1 | 2 \rangle + c_2^{\star}c_1\langle 2 | 1 \rangle = 1
$$
To find the coefficients $c_i$, we need to apply the full Hamiltonian to our LCAO ansatz:
$$
H|\Psi \rangle = \sum_i c_i H |i\rangle = E |\Psi\rangle 
$$
Taking the inner product with $|j\rangle$ gives:
$$
\langle j | E | \Psi\rangle = \sum_i c_i \langle j | H | i \rangle 
$$ 
but $\langle j | E |\Psi \rangle = c_j E$. Therefore, the above two equations can be written in the following matrix form:
$$
\begin{pmatrix}
\langle 1 | H | 1 \rangle & \langle 1 | H | 2 \rangle  \\
\langle 2 | H | 2 \rangle & \langle 2 | H | 1 \rangle  
\end{pmatrix}
\begin{pmatrix}
c_1 \\
c_2 
\end{pmatrix} = E \begin{pmatrix}
c_1 \\
c_2 
\end{pmatrix}
$$
We see that the **eigenvector-eigenvalue depends on only two paramters**, namely the so-called **on-site energies**, given by the matrix element $\langle i | H |i \rangle := E_0$, and the so-called **hopping integral**, given by the element $\langle i | H | j \rangle := -t $. Writing explicitly the on-site energy, we find that
$$
E_0 = \langle i | T+ \sum_k V_k | i \rangle = \varepsilon_0 + \langle i | V_j | i \rangle 
$$
with $j\neq i$, i.e. the on-site energy is the energy of the original atomic orbital (see the first eigenvector equation written for both $|1\rangle$ and $|2\rangle$) plus an energy shift due to the presence of the second potential well. For the hopping term, we have instead:
$$
-t = -\langle i | H | j \rangle = - \langle i | V_i | j \rangle
$$
again with $i\neq j$. 

Hence, we finally arrive at the following 2x2 Hamiltonian matrix:
$$
\boldsymbol{H} = \begin{pmatrix}
E_0 & -t \\
-t & E_0 \end{pmatrix}
$$
The diagonalization is trivial and gives:
$$
\det \begin{pmatrix}
E_0-\lambda & -t \\
-t & E_0-\lambda \end{pmatrix} = (E_0 - \lambda )^2 - t^2 = \lambda^2 - 2E_0\lambda + E_0^2-t^2
$$
i.e.
$$
\lambda_{1,2} = \frac{1}{2}\left( 2E_0 \pm \sqrt{ 4E_0^2 - 4E_0^2 + 4t^2}\right) = E_0 \pm t
$$

In this notebook, our goal is to solve a similar eigenvector-eigenvalue problem but in a more general form. Let indeed extend the above formulation to a more general case where we have a sequence of potential wells defining a 1D atomic chain: 

```python
def gaussian_well(x: np.array, a: float, b: float) -> np.array:
    return -np.exp(-(x-a)**2/b)
```

```python
x = np.linspace(-5, 5, 1000)
y = np.sum([gaussian_well(x, a, 0.01) for a in range(-4, 5)], axis=0)
```

```python
fig, ax = plt.subplots(figsize=(10, 2))

ax.plot(x, y, color="red", lw=2, alpha=0.5)
ax.set_xlabel("x")
ax.set_ylabel(r"$V(x)$")

plt.show()
```

Again, we suppose that each potential well supports a given number of energy states. In particular, we are interested in the low-energy states. At each atom position we introduce a basis set of functions $|i\rangle$, and therefore our task translates into finding the best linear combination of such basis functions. Let introduce the following extended notation: $|i_{\alpha}\rangle$ denotes the $i$th excited state of the $\alpha$ well. Therefore, we may write:
$$
|\Psi \rangle = \sum_{\alpha} \sum_i |i_{\alpha}\rangle
$$
In a matricial form, we may again represent the "molecular" state $|\Psi\rangle$ in terms of the coefficients of the linear combination, i.e.
$$
|\Psi\rangle = c_1 |0_0\rangle + c_2|1_0\rangle + c_3|0_1\rangle + \ldots \Longrightarrow \begin{pmatrix}
c_1 \\
c_2 \\
c_3 \\
\ldots \\
c_N
\end{pmatrix}
$$
It is worth clarifying the notation we are using. In order to keep the notation simple, we are using $c_i$ to denotes coefficients in the linear combination of states, but note that, if we assume to have two states (i.e. atomic orbitals) per well, then $c_1, c_2$ refers to the first well, $c_3, c_4$ to the second well, etc. Additionaly, we need to keep in mind that we are assuming all atoms to be of the same kind. 
The potential energy $V(x)$ will be the sum of all atomic potentials 
$$
V(x) = \sum_i V_i(x)
$$

To find the possible bound states of the molecular system, we need write the full Hamiltonian in the $|i_\alpha\rangle$ basis, as done above for the two-atom example; in other words, we need to find the matrix elements $\langle i_\alpha | H | j_\beta\rangle$. To this aim, we note that the full Hamiltonian can be re-written as:
$$
H = \sum_\alpha H_\alpha+ \hat{V}_\alpha
$$
where $H_\alpha$ is the Hamiltonian of the $\alpha$th-well and $\hat{V}_\alpha$ is the potential of all atoms *minus that of the atom $\alpha$ itself*, i.e. $\hat{V}_\alpha = \sum_{\gamma\neq \alpha} V_k$. Now, if $n = \alpha - \beta$ , i.e. the number of wells apart between $\alpha$ and $\beta$, we have a convenient way to write the Hamiltonian matrix element:
$$
\begin{align}
\langle i_\alpha | H | j_\beta\rangle &= \langle i_\alpha | H_{\alpha} + \hat{V}_{\alpha} | j_\beta\rangle  \\
&= \langle i_\alpha | E_j + \hat{V}_{\alpha} | j_\beta\rangle
\end{align}
$$
where $E_j$ is the energy of the $j$ state of the $\alpha$ well. Introducing a position basis, the above *bracket* dot product translates into
$$
\langle i_\alpha | H | j_\beta\rangle = \int_{-\infty}^{+\infty} \phi_{\alpha}^{\star} (x-na) (E_j - \hat{V}(x) ) \phi_{\beta}(x) dx
$$
where $a$ is the spacing between wells. It is interesting to note that the above multi-well expression actually depends just on single-well variables, therefore solving the eigenvalue problem for the single-well will automatically give us everything we need to find the energies of the multi-well system. Let's do it below with Python!

## Single-Well solution

Let first define a Python class with all methods needed to solve the single well problem. To this aim, let consider again the time-independent Schrodinger equation:
$$
\left( -\frac{d^2}{dx^2}+V(x) \right)\psi(x) = E\psi(x)
$$
We can discretize the space variable $x$ and convert the above eigenvalue-eigenvector equation to a matrix form as follows. We define a set of $N$ grid points $x_i$, with spacing $\Delta x$; the value of a function $\psi$ at the grid point $x_i$ will be $\psi_i \equiv \psi(x_i)$. The first-order differential operator in such a discrete form reads as
$$
\frac{d\psi}{dx} \approx \frac{\psi_{i+1}-\psi_i}{\Delta x} := g
$$
Therefore, the second-order differential operator may be approximated to
$$
\frac{dg}{dx} \approx  \frac{1}{\Delta x}(g_{i+1}-g_i) = \frac{1}{\Delta x^2} (\psi_{i+1}-\psi_{i}-\psi_{i}+\psi_{i-1})
$$
and the discrete Schrodinger equation reduces to
$$
\frac{2\psi_i - \psi_{i+1}-\psi_{i-1}}{\Delta x^2}+V_i\psi_i = E\psi_i
$$
We can assume the wavefunction $\psi$ to decay sufficiently fast outside the well, thus we can set the following boundary conditions: $\psi_0 =\psi_N = 0$. The above problem can then be re-written in the following matrix form:
$$
\begin{pmatrix}
\frac{1}{\Delta x^2}+V_1 & -\frac{1}{\Delta x^2} & 0 & \ldots & 0 \\
\frac{1}{\Delta x^2} & \frac{1}{\Delta x^2}+V_2 & -\frac{1}{\Delta x^2} &  \ldots & 0 \\
0 & \frac{1}{\Delta x^2} & \frac{1}{\Delta x^2}+V_3 & -\frac{1}{\Delta x^2} & \ldots  \\
0 & \ldots & \ldots & \ldots & -\frac{1}{\Delta x^2} \\
0 & 0 & 0 & -\frac{1}{\Delta x^2}  & \frac{1}{\Delta x^2}+V_{N-1}
\end{pmatrix}
\begin{pmatrix}
\psi_1 \\
\psi_2 \\
\psi_3 \\
\ldots \\
\psi_{N-1}
\end{pmatrix} = E \begin{pmatrix}
\psi_1 \\
\psi_2 \\
\psi_3 \\
\ldots \\
\psi_{N-1}
\end{pmatrix} 
$$

Notice that the first is a **tridiagonal matrix**, therefore we can use the `eigh_tridiagonal` method provided by the `scipy` package for the diagonalization.

```python
class SingleWell:
    def __init__(self, x: np.array, V: np.array, V0: float):
        self.x = x
        self.V = V
        self.V0 = V0
        self.dim = len(self.x)
        self.dx = np.diff(x)[0]
        self.main_diag, self.off_diag = self.__fill_hamiltonian_matrix()

    def __fill_hamiltonian_matrix(self) -> np.array:
        '''
        Private method to define the main and off diagonal elements of the Hamiltonian matrix
        '''
        main_diag = 2*np.ones(self.dim)/self.dx**2 + self.V
        off_diag = -np.ones(self.dim-1)/self.dx**2

        return main_diag, off_diag

    def __normalize_wavefunction(self, psi) -> np.array:
        '''
        Private method to normalize the eigenstates obtained by the Hamiltonian matrix normalization.
        '''
        norm = np.sqrt((np.abs(psi)**2).sum(axis=0)*self.dx)

        return psi/norm
    
    def diagonalize(self) -> Tuple[np.array]:
        '''
        Method to diagonalize the Hamiltonian matrix.

        :returns E: the numpy.array storing energies (eigenvalues)
        : returns psi: the numpy.array storing the eigenstates (eigenvectors)
        '''
        E, psi = eigh_tridiagonal(self.main_diag, self.off_diag, select="v", select_range=(-V0, 0))

        psi_norm = self.__normalize_wavefunction(psi)
        
        return E, psi_norm.T 
```

Below, we set basic parameters to define a finite potential well. In particular, we may set:
- $V_0 = 1.5$ eV 
- width of the well: $w = 0.50$ nm
- spacing: $a = 3*2w$

For simplicity, we divide by `dimless`, a parameter defined as $\hbar^2/(2mw^2)$, to make everything dimensionless.

```python
# Potential: eV
V0 = 1.5

# Plank constant in eV*s
hbar = 6.582112e-16 

# The electron mass in ev*s^2/m^2
m = 0.510999e6 / (3e8)**2 

# The width in m
w = 0.50e-9 / 2 

# Dimensionless conversion
dimless = hbar**2 / (2*m*w**2)

V0 = V0 / dimless

# dimensionless Spacing
a = 3*2
```

```python
# Defining the discrete grid 
N = 100000
x = np.linspace(-15,15,N)
```

As a first example, let consider a simple **Square Potential**:

```python
# Defining the square region to fill--> from -1 to +1, True
def create_boolean_square_region(x: np.array, x1: float, x2:float):
    return (x>=x1)*(x<=x2)

square_region = create_boolean_square_region(x, -1, 1)
square_region
```

```python
# Building the potential 
V = -V0*np.array(square_region, dtype=float)
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(x, V, lw=2, ls="--", color="red", alpha=0.5, label=r'$V(x)$')
ax.set_xlabel("x")
ax.set_ylabel("V(x)")
ax.legend()
plt.show()
```

```python
# Create Single Well Instance
sw = SingleWell(x, V, V0)
```

```python
# Diagonalizing the single well
E, psi = sw.diagonalize()
```

```python
psi.shape
```

Notice that in this case we have a single eigenstate, defined by a single numpy array of shape 1, 100000 where 10000 is the number of grid points defined above.

```python
x.shape[0]
```

## Single Band

Let now use the single-well solution to address the case of multiple square potential wells. First of all, in order to solve the integral numerically, we need to the *interpolate* the eigenstates to create a function. To this aim, we can make us of the `interp1d` method of the `scipy`package:

```python
psi.shape
```

```python
# Interpolation: again we cycle over available bound states, even if in this case we have just one state
psi_func = [ interpolate.interp1d(x, psi_i, bounds_error=False, fill_value=(0,0)) for psi_i in psi]
```

```python
psi_func
```

```python
# Defining a function that generate the potential well
def gen_V(x: np.array, V0: float, x1: float, x2: float) -> np.array:
    '''
    Function that generates a single potential well defined over a finite 1D-region bounded by a, b
    '''
    
    square_region = create_boolean_square_region(x, x1, x2)

    return -V0*np.array(square_region, dtype=float)
```

To arrange periodically multiple square potential, instead of passing the entire x grid, we can pass x%n, i.e. we transform the x array into an array storing the rest of the division by. In this way, at each multiple of n, we will have 0, with match the condition of the `create_boolean_square_region` function, if x1 and x2 are set to -1, 1 respectively.

```python
x%5
```

```python
plt.plot(x, gen_V(x%5, V0, -1, 1), lw=2, ls="--", color="red", alpha=0.5, label=r'$V(x)$')
```

However, we note that in this way, each potential is not correctly centered at the right point. Therefore, we need to apply a translation by half $n$:

```python
plt.plot(x, gen_V((x-2.5)%5-2.5, V0, -1, 1), lw=2, ls="--", color="red", alpha=0.5, label=r'$V(x)$')
```

Finally, we need to remove the potential at 0, according to the equation defined in the theory section, where we have seen that the potential to use must the periodic potential minus the potential of atom $\alpha$ itself.

```python
# Periodic potential
plt.plot(x, gen_V((x-2.5)%5-2.5, V0, -1, 1)-gen_V(x, V0, -1, 1), lw=2, ls="--", color="red", alpha=0.5, label=r'$V(x)$')
plt.show()
```

```python
# Wrap everything up into a function 
def V_periodic(x: np.array, a: float, V0: float, x1: float, x2: float):

    periodic_condition = (x-a/2)%a-a/2
    
    return gen_V(periodic_condition, V0, x1, x2) - gen_V(x,V0, x1, x2)
```

```python
x_extended = np.linspace(-20, 20, 10000)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x_extended, psi_func[0](x_extended)**2, color="royalblue", lw=2, label=r"$|0_0\rangle$")
ax.plot(x_extended, psi_func[0](x_extended-a)**2, color="darkorange", lw=2, label=r"$|0_1\rangle$")
ax.plot(x_extended, V_periodic(x_extended, a, V0, -1, 1)/V0, color="red", lw=2, ls="--", alpha=0.5)

ax.set_xlabel("x")
ax.legend()
plt.show()
```

```python
# Define a function to integrate the interpolated wavefunction
def integrate_state(psi_func: np.array, E: np.array, i1: int, i2: int, a: float, V0: float, n: int, x1: float, x2: float):

    lambda_func = lambda x: psi_func[i1](x)*psi_func[i1](x-n*a)*(E[i1]+V_periodic(x, a, V0, x1, x2))

    return integrate.quad(
        lambda_func, -20, 20, points = [n*a for n in range(-6, 6)], limit=1000)[0]
    
```

```python
# Define number of states 
N_states  = len(E)
N_states
```

```python
# Define number of wells
N_wells = 50
```

```python
# Initiliazing an Hamiltonian matrix, of dimension N states x N wells
H = np.zeros((N_wells*N_states, N_wells*N_states))
```

```python
H.shape
```

Now, we should compute each Hamiltonian matrix element applying the `integrate_state` function. First of all, let consider the following multiplication:

```python
len(N_wells*psi_func)
```

The above multiplication repeat the interpolated wavefunction for a number of times equal to the number of wells. In this case, the list dimension is 50 because `psi_func` is a numpy array storing just a single eigenstate (i.e. with shape 1, 10000, where 10000 is the number of grid points). Therefore, the element $i$ of the above list is the represent the state of the $i$ well. If `psi_func` were a numpy array of shape (2, N) i.e. two rows representing two different state, the ground state and the excited state, then the above multiplication would give a list of `N_wells*2` dimension. In such case, the first two elements of the list would be the ground and excited state of the first well, the 3r and 4th elements would be the ground and excited state of the second well, and so on.

```python
# Let see how we can fill
test_M  = np.zeros((5, 5))
for i1 in range(test_M.shape[0]):
    for i2 in range(i1, test_M.shape[0]):
        test_M[i1, i2] = 1
```

```python
test_M
```

With the above combination of two for loops, we can see that we can fille the off-diagonal elements. Let us see now how we can correctly take the indices of the potential wells:

```python
test_states = 1
for i1 in range(test_M.shape[0]):
    for i2 in range(i1, test_M.shape[0]):
        print(f"i1={i1}, i2={i2}, {i1//test_states}, {i2//test_states}")
```

```python
psi_band = N_wells*psi_func
E_band = N_wells*list(E)
```

```python
len(psi_band), len(E_band)
```

```python
# Filling the upper tridiagonal part of the Hamiltonian matrix
for i1 in range(H.shape[0]):
    for i2 in range(i1, H.shape[0]):
        id_x_well1 = i1//N_states 
        id_x_well2 = i2//N_states
        H[i1, i2] = integrate_state(
            psi_func=psi_band, E=E_band,  i1=i1, i2=i2, a=a, V0=V0, n=id_x_well2-id_x_well1, x1=-1, x2=1)
```

```python
# Filling the lower tridiagonal part of the matrix 
H += H.T
```

```python
H
```

```python
#  Now fill the main diagonal
main_diag = np.array(E_band)
H[np.diag_indices(H.shape[0])] = main_diag
```

Let try to visualize the Hamiltonian matrix in a plot with a color map to see if we have correctly filled it:

```python
plt.imshow(H, cmap="Blues")

plt.colorbar()
plt.show()
```

We can finally diagonalize the filled Hamiltonian matrix and plot the resulting eigenstates:

```python
eigenenergies, eigenvectors = np.linalg.eigh(H)
```

```python
fig, ax = plt.subplots(figsize=(4, 6))

ax.eventplot(dimless*eigenenergies, orientation="vertical", colors="royalblue", lw =1)
ax.set_ylabel("Energies")
ax.set_title("Single Band")
plt.show()
```

We can observe the emergence of a **band**. The periodic arrangement of potential wells gives rise to a continuous spectrum of possible energies for the electron in the system. This spectrum is not exactly continuous in our simulation because we have limited the number of wells for computational reasons. However, if we could simulate an infinite systema system with an infinite number of potential wells (atoms) and the same translational symmetry used in our construction of the total potential energywe would see the energy levels becoming so close to each other that they form a continuous spectrum. This behavior is typical of metals, and it is this unique electronic structure that gives metals their interesting properties, such as electrical conductivity and more.

## Two Bands

Let's make our simulation a little more complex. Specifically, let's increase the depth of our potential well so that it can also support a first excited state:

```python
# Depth in ev 
V0 = 5 # ev
w = 0.35e-9 / 2 # m

# dimensionless
dimless = hbar**2 / (2*m*w**2)
V0 = V0 / dimless
a = 3*2
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(x, gen_V(x, V0, -1, 1), lw=2, ls="--", color="red", alpha=0.5, label=r'$V(x)$')
ax.set_xlabel("x")
ax.set_ylabel("V(x)")
ax.legend()
plt.show()
```

```python
V = gen_V(x, V0, -1, 1)
```

```python
sw2 = SingleWell(x, V, V0)
```

```python
E2, psi2 = sw2.diagonalize()
```

We can see that in this case we have two bound states, corresponding to two different energies.

```python
fig, ax = plt.subplots(1, 2, figsize=(12,4))

colors = ["royalblue", "blueviolet"]

# Plot the wave functions
for i, psi_i in enumerate(psi2):
    color = colors[i % len(colors)]  # Cycle through colors if more than 2 psi
    ax[0].plot(x, psi_i, lw=2, color=color)

for i, E_i in enumerate(E2):
    color = colors[i % len(colors)]  # Cycle through colors if more than 2 E
    ax[1].plot(x, np.full(x.shape[0], E_i), lw=2, color=color)

for axi in ax:
    axi.set_xlabel("x")

ax[0].set_title("Boun states of the single well")
ax[1].set_title("Bound states energies")

plt.show()
```

```python
# let interpolate the numerical wavefunction
psi2_func = [interpolate.interp1d(x, psi, bounds_error=False, fill_value=(0,0)) for psi in psi2]
```

In the plot below, we display the ground state of the first well, together with the first excited state of the second potential well:

```python
x_extended = np.linspace(-20, 20, 10000)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x_extended, psi2_func[0](x_extended)**2, color="royalblue", lw=2, label=r"$|0_0\rangle$")
ax.plot(x_extended, psi2_func[1](x_extended-a)**2, color="goldenrod", lw=2, label=r"$|1_1\rangle$")
ax.plot(x_extended, V_periodic(x_extended, a, V0, -1, 1)/V0, color="red", lw=2, ls="--", alpha=0.5)

ax.set_xlabel("x")
ax.legend()
plt.show()
```

As before, we fill the Hamiltonian matrix with the usual integral, then we diagonalize the matrix to compute the eigenenergies and eigenstates:

```python
N_wells = 50
N_states = len(E2)
H = np.zeros((N_wells*N_states, N_wells*N_states))
```

```python
psi2_band = N_wells*psi2_func
E2_band = N_wells*list(E2)
```

```python
for i1 in range(H.shape[0]):
    for i2 in range(i1, H.shape[0]):
        id_x_well1 = i1//N_states
        id_x_well2 = i2//N_states
        H[i1,i2] = integrate_state(psi2_band, E2_band, i1, i2, 
                                    a=a, V0=V0, n=id_x_well2-id_x_well1, x1=-1, x2=1)
H += H.T
```

```python
# fill main diag
main_diag = np.array(list(E2)*N_wells)
H[np.diag_indices(H.shape[0])] = main_diag
```

```python
plt.imshow(H, cmap="Blues")

plt.colorbar()
plt.show()
```

```python
eigenenergies, eigenvectors = np.linalg.eigh(H)
```

We finally plot the eigenenergies and see that in this case we have **two bands**:

```python
fig, ax = plt.subplots(figsize=(4, 6))

ax.eventplot(dimless*eigenenergies, orientation="vertical", colors="royalblue", lw =1)
ax.set_ylabel("Energies")
ax.set_title("Two Band")
plt.show()
```

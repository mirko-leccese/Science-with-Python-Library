```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_electrons = 100  # Number of electrons
field_strength = 1e-3  # Electric field (V/m)
electron_charge = 1.6e-19  # Charge of an electron (C)
electron_mass = 9.11e-31  # Mass of an electron (kg)
relaxation_time = 1e-14  # Relaxation time (s)
time_step = 1e-15  # Time step (s)
num_steps = 1000  # Number of simulation steps

# Initialize electron positions and velocities
positions = np.zeros(num_electrons)
velocities = np.random.normal(0, 1e5, num_electrons)  # Random initial velocities (m/s)

# Drift velocity
drift_velocity = -(electron_charge * field_strength * relaxation_time) / electron_mass

# Time evolution
position_history = []
for step in range(num_steps):
    # Update positions based on current velocity
    positions += velocities * time_step
    
    # Apply drift due to the electric field
    velocities += (electron_charge * field_strength / electron_mass) * time_step
    
    # Apply random scattering
    scatter = np.random.rand(num_electrons) < (time_step / relaxation_time)
    velocities[scatter] = np.random.normal(0, 1e5, np.sum(scatter))
    
    # Store positions for visualization
    position_history.append(positions.copy())

# Convert position history to a numpy array
position_history = np.array(position_history)

# Plotting the simulation
plt.figure(figsize=(10, 6))
for i in range(num_electrons):
    plt.plot(np.arange(num_steps) * time_step, position_history[:, i], alpha=0.6)
plt.title("Electron Motion in the Drude Model")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.grid()
plt.show()
```
